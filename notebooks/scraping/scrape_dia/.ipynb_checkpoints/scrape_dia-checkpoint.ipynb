````python
def contract_expand():
    try:
        #Opens Categories
        driver.find_element(By.CSS_SELECTOR, "button[aria-label='categories list menú button']").click()

    except:
        #Contracts/closes categories
        driver.find_element(By.CSS_SELECTOR, "span[data-test-id='categories-list-modal-cross']").click()
````

````python
# generamos una funcion porque la vamos a usar mucho 
"""Contra / Expande una categoria. 
Hay que contraer/expandir una categoria para seleccionar otra"""

def contract_expand_categ(c):
    
    c.find_element(By.CSS_SELECTOR, "span[data-test-id='category-item-symbol-icon']").click()
````

## ENTRAMOS AL DIA Y CERRAMOS LAS COOKIES;

````python
def start_driver_dia():

    # Start the driver with the options
    driver = webdriver.Chrome()

    # Navigate to the login page
    link = f'https://www.{supermercado}.es/'
    driver.get(link)

    time.sleep(3)
    #Cerramos la Cookies 
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(),'Rechazar todas')]"))).click()
    
    return driver
````

````python
def create_dictionary():

    #Abrimos las categorias: 
    contract_expand()

    # EXTRAEMOS LAS CATEGORIAS: 
    category_links = driver.find_elements(By.CSS_SELECTOR, "div[data-test-id='category-item']")

    cat_subcat_dict = {}

    for c in category_links:

        print(c.text)

        #EXPANDIMOS CATEGORIA SELECCIONADA:
        time.sleep(1)
        contract_expand_categ(c)

        #WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, "div[data-test-id='sub-category-item']")))

        time.sleep(2)
        #Sacamos SUBCATEOGRIAS: 
        subcategory_links = driver.find_elements(By.CSS_SELECTOR, "div[data-test-id='sub-category-item']")

        cat_subcat_dict[c.text] = [i.text for i in subcategory_links]
        
    return cat_subcat_dict
````

````python


#Click on Categories Button(Wait til its visible)
WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.LINK_TEXT, "Categorías"))).click()

#Save all categories into a variable:
category_links = driver.find_elements(By.CSS_SELECTOR, "span[class='category-menu__header']")

for c in category_links:
    category = c.text
    #print(category)
    
    #Enter the category:
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//label[text()='{category}']"))).click()
    
    #Target subcategories:
    subcategory_links = driver.find_elements(By.CSS_SELECTOR, ".category-item__link")
        
    for s in subcategory_links:
        subcategory = s.text
        print(subcategory)
        
        #Enter a subcategorie:
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//button[text()='{subcategory}']"))).click()
        #print("subclicada")
        
        # Wait for the products to load and saves them
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".product-cell")))

        #Extrae todos los productos(cajita que vimos )
        product_cells = driver.find_elements(By.CSS_SELECTOR, "div[data-test='product-cell']")

        for p in product_cells:
            product = p
            #print(product)
            
            nueva_fila = {
                'categoria':category , 
                'subcategoria': subcategory, 
                'descripcion': product.find_element(By.CSS_SELECTOR, "h4[data-test='product-cell-name']").text, 
                "unidades": product.find_element(By.CSS_SELECTOR, "div[class='product-format product-format__size--cell']").text, 
                "precio": product.find_element(By.CSS_SELECTOR, "div[class='product-price']").text
                }


            df = pd.concat([df, pd.DataFrame([nueva_fila])], ignore_index=True)
            #print("row inserted")
````

````python
def export_to_csv(data_frame, file_name):
    """
    Export a DataFrame to a CSV file.

    Parameters:
        - data_frame: pandas DataFrame
        - file_name: str, name of the CSV file (without the extension)
    """
    #print date stamp - datetime.datetime.now()
    
    t_stamp = str(datetime.datetime.now()).split(".")[0].replace(" ","_")
    
    file_path = f"{file_name}_{t_stamp}.csv"
    
    data_frame.to_csv(file_path, index=False)
    print(f"DataFrame successfully exported to {file_path}")
````
